# Optional Parameters
- `Optional[str]` is shorthand for `str | None` in modern Python.
- It does not mean the argument is optional by itself.
- A parameter is optional only if it has a default value, like `= None`.

```python
from typing import Optional

def greet(name: Optional[str] = None) -> str:
    if name is None:
        return "Hello, world!"
    return f"Hello, {name}!"
```

But, these are different

```python
def f(x: Optional[int]):      # x must still be passed, but can be int or None
    ...

def g(x: Optional[int] = None):  # x may be omitted, and if provided can be int or None
    ...
```
